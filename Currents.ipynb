```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import math
```

## Analysis of ion currents

Read all files with current

```python
dt = 0.1 # timestep
```

```python
data = {}
files = []
for filename in os.listdir('.'):
    if filename.startswith('curr'):
        values = pd.read_csv(filename, sep = ' ', names=['t', 'I'])
        values = values[(values['t'] > 60) & (values['t'] < 190)]
        values['I_cum'] = np.cumsum([values['I']]) * dt
        data[filename[:-6]] = values
        files.append(filename[:-6])
        
        name = filename[:-4].split('_')
        current_type = name[0]
        mutant = name[1]
        voltage = int(name[2][:-2])
        print(filename[:-6])
        display(data[filename[:-6]].head(3))
        print(len(values))
```

```python
print(files)
```

Get average current by making linear fit of cumulutive sum

```python
# Names of files
mutants = [s.split('_')[1] + '_' + s.split('_')[2] for s in files]

# Table with currents
summary = pd.DataFrame(columns=['mut', 'V', 'curr', 'curr_err',
                                'currCLA', 'currCLA_err', 'currPOT', 'currPOT_err'],
                       index = mutants)
for f in files:
    plt.plot(data[f]['t'], data[f]['I_cum'])
    dt = 0.1
    p, cov = np.polyfit(data[f]['t'], data[f]['I_cum'], 1, cov = True)
    errorbars = np.sqrt(np.diag(cov))
    plt.plot(data[f]['t'], [p[0] * t + p[1] for t in data[f]['t']])
    plt.xlabel('Time, ns')
    plt.ylabel('Cumulative current')
    
    name = f.split('_')
    curr_type = name[0]
    mutant = name[1]
    voltage = name[2]
    summary.loc[mutant + '_' + voltage, 'mut'] = mutant
    summary.loc[mutant + '_' + voltage, 'V'] = int(voltage)
    summary.loc[mutant + '_' + voltage, curr_type] = p[0]
    summary.loc[mutant + '_' + voltage, curr_type + '_err'] = errorbars[0]
```

```python
plt.plot(data['curr_K238R_150']['t'], data['curr_K238R_150']['I_cum'])
```

```python
summary = summary.drop_duplicates()
exp_error = 0.009
for col in ['curr_err', 'currPOT_err', 'currCLA_err']:
    summary[col + '_lr'] = summary[col]
    summary[col] = summary[col].apply(lambda x: math.sqrt(x ** 2 + exp_error ** 2))
summary = summary.sort_index()
summary
```

There are some cases with very small negative current, probably we should set them to zero.  

In general we see that the chloride current is larger than potassium.

```python
#summary = summary[summary['V'] >= 0]
plt.figure(figsize=(7, 5))

plt.errorbar(summary.index, summary.currPOT, yerr=summary.currPOT_err, label='Potassium current', 
             linestyle='None', marker='o')
plt.errorbar(summary.index, summary.curr, yerr=summary.curr_err, label="Total current", linestyle='None', marker='o')
plt.errorbar(summary.index, summary.currCLA, yerr=summary.currCLA_err, label='Chloride current', linestyle='None', marker='o')
plt.hlines(0, 'K238A_-150', 'Wt_450')

plt.xticks(rotation='vertical')
plt.ylabel('Current')
plt.legend()
```
